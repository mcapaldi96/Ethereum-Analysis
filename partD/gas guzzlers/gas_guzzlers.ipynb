```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Average Gas

```python
colnames=['Date','Average Gas (Wei)']
aveGas = pd.read_csv('./Average_Gas', names=colnames)
# cleaning the columns
aveGas['Date']= aveGas['Date'].str.strip('()')
aveGas['Average Gas (Wei)']= aveGas['Average Gas (Wei)'].str.strip('()')
# converting gas back into a float
aveGas['Average Gas (Wei)'] = aveGas['Average Gas (Wei)'].astype(float)
aveGas
```

```python
fig, ax = plt.subplots(figsize = (15,6))    
fig = sns.lineplot(x='Date', y='Average Gas (Wei)', data=aveGas, 
                  estimator = sum, ci = None, ax=ax, color='Blue')

x_dates = aveGas['Date']
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
plt.ylabel('Average Gas (Wei)')
plt.title('Average Gas Price vs Time')
plt.show()
```

## Difficulty and Gas price over time

```python
colnames=['Date','Difficulty','Gas (Wei)']
df = pd.read_csv('./Difficulty_Time', names=colnames)

# cleaning the columns
df['Date']= df['Date'].str.strip('(')
df['Difficulty']= df['Difficulty'].str[2:]
df['Difficulty']= df['Difficulty'].astype(float) 
# converting gas back into a float
#df['Difficulty'] = df['Difficulty'].astype(float)
df['Gas (Wei)']= df['Gas (Wei)'].str.strip('))')
df['Gas (Wei)']= df['Gas (Wei)'].astype(float)
# converting gas back into a float
```

```python
df
```

```python
fig, ax = plt.subplots(figsize = (15,6))    
fig = sns.lineplot(x='Date', y='Difficulty', data=df, 
                  estimator = sum, ci = None, ax=ax, color='Red')

x_dates = df['Date']
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
plt.ylabel('Difficulty')
plt.title('Difficulty vs Time')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,6))    
fig = sns.lineplot(x='Date', y='Gas (Wei)', data=df, 
                  estimator = sum, ci = None, ax=ax, color='Green')

x_dates = df['Date']
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
plt.ylabel('Gas fee (Wei)')
plt.title('Gas fees vs Time')
plt.show()
```
