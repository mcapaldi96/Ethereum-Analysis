```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# converting count transaction text file to a csv file
read_file = pd.read_csv (r'/Users/mikem1/Documents/Masters/ECS765P Big Data Processing/Ethereum Coursework/partA/CountTransactions')
read_file.to_csv (r'/Users/mikem1/Documents/Masters/ECS765P Big Data Processing/Ethereum Coursework/partA/CountTransactions.csv', index=None)
```

```python
# converting average transaction text file to a csv file
read_file = pd.read_csv (r'/Users/mikem1/Documents/Masters/ECS765P Big Data Processing/Ethereum Coursework/partA/averageTransactions')
read_file.to_csv (r'/Users/mikem1/Documents/Masters/ECS765P Big Data Processing/Ethereum Coursework/partA/averageTransactions.csv', index=None)
```

```python
# reading in csv file and creating the dataset for transaction count
colnames=['Date','No. of Transactions'] 
df_ctrans = pd.read_csv('./CountTransactions.csv', names=colnames, header=None, delimiter='\t').sort_values('Date')
df_ctrans['Date'] = pd.to_datetime(df_ctrans['Date'])
```

```python
# reading in csv file and creating the dataset for average transactions
colnames=['Date','Average Value of Transactions'] 
df_avetrans = pd.read_csv('./averageTransactions.csv', names=colnames, header=None, delimiter='\t').sort_values('Date')
df_avetrans['Date'] = pd.to_datetime(df_ctrans['Date'])
```

```python
# plotting transaction count graph
fig, ax = plt.subplots(figsize = (15,6))    
fig = sns.barplot(x='Date', y='No. of Transactions', data=df_ctrans, 
                  estimator = sum, ci = None, ax=ax, color='Blue')

x_dates = df_ctrans['Date'].dt.strftime('%Y-%m').sort_values().unique()
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
plt.ylabel('No. of Transactions (tens of millions)')
plt.title('Total number of transactions per month')
plt.show()
```

```python
# plotting average transactions graph
fig, ax = plt.subplots(figsize = (15,6))    
fig = sns.barplot(x='Date', y='Average Value of Transactions', data=df_avetrans, 
                  estimator = sum, ci = None, ax=ax, color='Red')

x_dates = df_avetrans['Date'].dt.strftime('%Y-%m').sort_values().unique()
ax.set_xticklabels(labels=x_dates, rotation=45, ha='right')
plt.title('Average value of transactions per month')
plt.show()
```
